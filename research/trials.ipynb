## Using box python package to manupulate the dictionaries

```python
from box import ConfigBox

dict = {
    "key1" : "value1",
    "key2" : "value2"
}
print(dict)
```

```python
dict["key1"]
```

```python
boxDict = ConfigBox(dict)
boxDict.key1
```

## Using of ensure annotation

```python
def function1(x:int, y:int):
    return x*y
```

```python
function1(x=4, y=6)
```

```python
function1(x=4, y="6")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def function2(x:int, y:int):
    return x*y

try:
    function2(x=4, y="6")
except:
    print("Error in type")
```
